### PCA

```python
def pca(X, dims=10):
    (n, d) = X.shape
    X = X - np.tile(np.mean(X, 0), (n, 1))
    (l, M) = np.linalg.eig(np.dot(X.T, X))
    Y = np.dot(X, M[:, 0:dims])
    return Y
```

```python
X_pca = pca(X)
X_pca[:5]
```

### KMeans

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=8, random_state=42).fit(X_pca[:200000, :])
y = kmeans.labels_
pre = kmeans.predict(X_pca[200000:, :])
kmeans.score(X_pca[200000:, :])
pre += 1
pre
```
