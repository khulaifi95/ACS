```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/home/kevinxu95/Data/order.csv')
```

```python
df = df[:500000]
df = df[['sortorderid', 'productcode', 'sortquantity']]
```

```python
df = df.pivot_table(index='sortorderid', columns='productcode', values='sortquantity')
df = df.fillna(0)
```

```python
df['Total']= df.sum(axis=1)
cols = df.columns.tolist()
cols = cols[-1:] + cols[:-1]
df = df[cols]
```

```python
df.head()
```

```python
df.describe()
```

```python
df1 = df[df['Total']!=1]
df2 = df[df['Total']==1]
```

```python
not_int = []
for idx, i in enumerate(df1['Total']):
    if i != int(i):
        not_int.append(idx)

df1 = df1.drop(df1.index[not_int])
df1 = df1.iloc[:,1:]
# df1 = df1.drop(df1.columns[df1.max(axis=0)==1], axis=1)
df1 = df1.iloc[:500]
```

```python
df1.describe()
```

```python
df1.shape
```

```python
from sklearn.decomposition import SparsePCA

Y = df1.values
transformer = SparsePCA(n_components=5, random_state=0)
transformer.fit(Y)

Z = transformer.transform(Y)
Z.shape
# np.mean(transformer.components_ == 0)
```

```python
pd.DataFrame(Z).to_csv("SPCA_order.csv", header=False, index=False)
```

```python
pd.DataFrame()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=0).fit(Z)
np.unique(kmeans.labels_, return_counts=True)
```
